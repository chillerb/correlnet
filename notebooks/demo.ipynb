```python
import numpy as np
import matplotlib.pyplot as plt

from correlnet.segments import SegmentPlotter
```

```python
x, y, x_end, y_end = np.random.normal(size=(4,5))
```

```python
from matplotlib.collections import LineCollection

fig, ax = plt.subplots()
plotter = SegmentPlotter()
ax = plotter.plot(x, y, x_end, y_end, linewidths=[1,2,3,4,5], colors=np.array([1,2,3,4,5]), ax=ax)
ax.autoscale()
```

```python
from matplotlib.colors import Normalize


fig, ax = plt.subplots()


lines = LineCollection([[(x0, y0), (x1, y1)] for x0, y0, x1, y1 in zip(x,y,x_end, y_end)], cmap="viridis", norm=Normalize())

# lines.set_array([1,2,3,4,5])
#lines.set_colors([1,2,3,4,5])
# lines.set_array(["red"] * 5)
#lines.set_colors(["red"] * 5)
# lines.set_array(["r"] * 5)
ax.add_collection(lines)
ax.autoscale()
```
